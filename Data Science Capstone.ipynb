```python
#Assign Not assigned 'Neighbourhood' value with 'Borough' Values
neighborhoods.loc[(neighborhoods['Neighbourhood'] == 'Not assigned'),'Neighbourhood']= neighborhoods['Borough']
```

```python
df = pd.read_csv('http://cocl.us/Geospatial_data', header=0)
df = df.rename(columns={'Postal Code': 'Postcode'})
```

```python
new_df = pd.merge(neighborhoods,df, on='Postcode')
new_df
```

create map to tornto

```python
address = 'Toronto, ON'

geolocator = Nominatim(user_agent="trt_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python

map_toronto = folium.Map(location=[latitude, longitude],zoom_start=10)
# add markers to map
for lat, lng, borough, neighborhood in zip(new_df['Latitude'], new_df['Longitude'], new_df['Borough'], new_df['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
